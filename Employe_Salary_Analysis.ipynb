## üìñ Project Overview
This project presents an analysis of employee salaries across different job roles, experience levels, employment types, and geographical locations. The dashboards were created using **Power BI** and are aimed at providing actionable insights into salary trends. The analysis helps HR departments, recruiters, and job seekers understand salary distributions and how various factors influence compensation.

---

## üîç Objectives
- To analyze salary trends based on job roles, experience levels, and employment types.
- To explore how geographical regions and remote work influence employee salaries.
- To identify salary disparities by company size and region.

## üìú Methodology
1. **Data Cleaning**:
   - Removed null values and duplicates in Excel.
   - Standardized column names and formats.
   - Created measures in Power BI for salary analysis, including:
     - **Max Salary**
     - **Min Salary**
     - **Avg Salary**
     - **Job Categories** (e.g., Software Engineer, Data Scientist, etc.)
2. **Data Analysis**:
   - Calculated key performance indicators (KPIs) such as **max salary**, **min salary**, **average salary**, and **salary by job category**.
   - Filtered data by region, experience level, and employment type.

3. **Visualization**:
   - Created interactive dashboards with bar charts, pie charts, and map visualizations.
   - Dashboard 1: **Employee Salary Overview** ‚Äì Shows salary distribution by job title, experience level, and employment type.
   - Dashboard 2: **Regional and Remote Work Analysis** ‚Äì Compares salaries by region and employment type (remote vs. non-remote).

---

## üõ†Ô∏è Features
- **Interactive Filters**: Allows users to filter data by job title, experience level, and region.
- **Key Visualizations**:
  - **Bar Charts**: Visualize salary distribution by job title and experience level.

  - **Pie Chart**: Displays salary distribution by region.
